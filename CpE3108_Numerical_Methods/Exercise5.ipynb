```python
#GLOBAL DECLARATIONS FOR NUMBER 2
import math

def f(x):
    function  = math.cos(x) - (3/(x)**5)

    if (math.isnan(function)==True):
        return 1
    else:
        return function
```

```python
#NUMBER 2: TRAPEZOIDAL METHOD
def trapezoid_fx2(a,b,n):
    fa = f(a)
    fb = f(b)
    h = (b - a) / n
    summation = 0
    for y in range (1, n):
        summation += f(a + y*h)
    print(summation)
    print(2*summation)
    f_trapezoid = (h/2) * (fa + 2*summation + fb)
    return f_trapezoid

print(round(f(1),6))
print(round(f(4),6))
print(trapezoid_fx2(1,4,4))
```

```python
#NUMBER 2: SIMPSON METHOD

def simpson_fx(a,b,n):
    fa = f(a)
    fb = f(b)
    h = (b - a) / n
    summation_odd = 0
    summation_even = 0
    for i in range(1, n, 2):
        summation_odd += f(a + i*h)
        print(summation_odd)
    for j in range(2, n-1, 2):
        summation_even += f(a + j*h)
        print(summation_even)
    f_simpson = (h/3) * (fa + 4*summation_odd + 2*summation_even + fb)
    return f_simpson 

print(round(f(1),6))
simpson_fx(1,4,4)
print(round(f(4),6))
```
